# Univariate Analysis

```python
def UVA(data,var1,var2,var3):
    plt.subplot(131)
    sns.countplot(data[var1])
    plt.xlabel(var1)
    plt.ylabel('Frequency')
    
    plt.subplot(132)
    sns.countplot(data[var2])
    plt.xlabel(var2)
    plt.ylabel('Frequency')
    
    plt.subplot(133)
    sns.countplot(data[var3])
    plt.xlabel(var3)
    plt.ylabel('Frequency')
```

# Missing Values

```python
data.isnull().sum()
```

```python
avg_culmen_length=data['culmen_length_mm'].mean()
avg_culmen_depth=data['culmen_depth_mm'].mean()
avg_flipper_length=data['flipper_length_mm'].mean()
avg_body_mass=data['body_mass_g'].mean()
mode_sex=data['sex'].mode()[0]
```

```python
print("Average of culmen length:", avg_culmen_length)
print("Average of culmen depth:", avg_culmen_depth)
print("Average of flipper length:", avg_flipper_length)
print("Average of body mass:", avg_body_mass)
print("Mode of sex:", mode_sex)
```

```python
data['culmen_length_mm']=data['culmen_length_mm'].fillna(avg_culmen_length)
data['culmen_depth_mm']=data['culmen_depth_mm'].fillna(avg_culmen_depth)
data['flipper_length_mm']=data['flipper_length_mm'].fillna(avg_flipper_length)
data['body_mass_g']=data['body_mass_g'].fillna(avg_body_mass)
data['sex']=data['sex'].fillna(mode_sex)
```

```python
data.isnull().sum()
```

The missing values were found and replaced according to their datatypes. For float variables, missing values were replaced with their mean value and for object variable, missing values were replaced with their mode

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(train_x,train_y)
y_pred=nb.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,test_y))
```

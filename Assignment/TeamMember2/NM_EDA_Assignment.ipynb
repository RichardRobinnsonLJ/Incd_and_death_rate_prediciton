# Univariate Analysis

```python
plt.figure(figsize=(22, 6))

#number of bedrooms
plt.subplot(131)
sns.countplot(data['number of bedrooms'])
plt.xlabel('number of bedrooms')

#number of bathrooms
plt.subplot(132)
sns.countplot(data['number of bathrooms'])
plt.xlabel('number of bathrooms')
plt.xticks(rotation=90)

#number of floors
plt.subplot(133)
sns.countplot(data['number of floors'])
plt.xlabel('number of floors')
```

It is observed that that most houses have 3 bedrooms and also most houses have 1 floor

```python
plt.figure(figsize=(22, 6))

#living area
plt.subplot(121)
sns.countplot(data['living area'])
plt.xlabel('living area')

#lot area
plt.subplot(122)
sns.countplot(data['lot area'])
plt.xlabel('lot area')
```

```python
plt.figure(figsize=(22, 6))

#living area renov
plt.subplot(121)
sns.countplot(data['living_area_renov'])
plt.xlabel('living area renov')

#lot area renov
plt.subplot(122)
sns.countplot(data['lot_area_renov'])
plt.xlabel('lot area renov')
```

living area, living_area_renov, lot area and lot_area_renov are all having huge number of varying data points and hence it

```python
plt.figure(figsize=(22, 6))

#waterfront present
plt.subplot(121)
sns.countplot(data['waterfront present'])
plt.xlabel('waterfront present')

#number of views
plt.subplot(122)
sns.countplot(data['number of views'])
plt.xlabel('number of views')
```

Most of the houses doesn't have a waterfront present,
Number of views is 0 for most of the houses

```python
plt.figure(figsize=(22, 6))

#condition of the house
plt.subplot(221)
sns.countplot(data['condition of the house'])
plt.xlabel('condition of the house')

#grade of the house
plt.subplot(222)
sns.countplot(data['grade of the house'])
plt.xlabel('grade of the house')

#Area of the house(excluding basement)
plt.subplot(223)
sns.countplot(data['Area of the house(excluding basement)'])
plt.xlabel('Area of the house(excluding basement)')

#Area of the basement
plt.subplot(224)
sns.countplot(data['Area of the basement'])
plt.xlabel('Area of the basement')
```

For most houses condition is 3 and for most houses grade is 7, Area of the house and basement seems to be having huge number of 
datapoints

```python
plt.figure(figsize=(22,6))

#Distance from the airport
plt.subplot(121)
sns.countplot(data['Distance from the airport'])
plt.xlabel('Distance from the airport')

#Number of schools nearby
plt.subplot(122)
sns.countplot(data['Number of schools nearby'])
plt.xlabel('Number of schools nearby')
```

Both the distance from the airport and number of schools datapoint doesn't seem to have much difference in their

# Bivariate Analysis

```python
def BVA(data,var):
    sns.barplot(data=data,x=var,y='Price')
```

```python
BVA(data,'number of bedrooms')
```

```python
BVA(data,'number of bathrooms')
plt.xticks(rotation=90)
```

```python
BVA(data,'number of floors')
```

The price increases with the increase in number of bathrooms and number of floors

```python
BVA(data,'waterfront present')
```

The price is high when there is waterfront

```python
BVA(data,'number of views')
```

The price increases with the increase in number of views

```python
BVA(data,'condition of the house')
```

```python
BVA(data,'grade of the house')
```

The price increases with increase in condition and grade of the house

```python
BVA(data,'Number of schools nearby')
```

```python
BVA(data,'Distance from the airport')
plt.xticks(rotation=90)
```

As already guessed, the number of schools nearby and distance from the airport doesn't affect price in a huge way

# Multivariate Analysis

```python
#Isolating the variables that affected price significantly in bivariate analysis
data_price_house_size=['number of bedrooms','number of bathrooms','number of floors','waterfront present','number of views',
                      'condition of the house','grade of the house']
data1=pd.DataFrame()
for i in data_price_house_size:
    data1[str(i)]=data[i]
House_size=data1.columns
```

```python
data1['Price']=data['Price']
```

```python
sns.pairplot(data1,vars=House_size,hue ='Price',plot_kws={'alpha':0.1})
plt.show()
```
